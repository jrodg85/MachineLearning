# SVM

uno de los algoritmos que mejores resultados, en general, obtiene para diferentes tipos de problemas son los SVM. Se consideran, junto a Random Forest (RF) uno de los algoritmos estado del arte que deben probarse para analizar cualquier problema. Son, por tanto, un punto de partida necesario en todo análisis.

Son algoritmos capaces de realizar una clasificación lineal (los datos se pueden separar linealmente) o no lineal tanto en problemas de clasificación, como regresión o detección de outliers.

Se dice que ellos que están diseñados para problemas particularmente complejos, pero de pequeño tamaño.

Si los datos son linealmente separables, el SVM básicamente tratará de separar con una línea (como siempre, vamos a pensar en dos dimensiones) los ejemplos de cada una de las clases.

## Quitando restricciones y añadiendo flexibilidad al modelo

si no fuésemos flexibles a la hora de permitir algún error, nos encontaríamos con el problema de que el modelo estaría muy limitado en cuanto a las posibles soluciones que es capaz de encontrar y, además, posiblemente dichas soluciones no serían buenas. La idea es permitir que el modelo tenga mayor capacidad para *jugar* con el problema. Esto se logra en un SVM con el hiperparámetro *C*, que permite por un lado que la separación sea la mayor posible y por el otro, que se produzcan violaciones de las restricciones impuestas al margen.

## ¿Y en regresión?

muy fácil, podemos pensar que es el objetivo inverso. Antes nos interesaba que la separación fuese la máxima (cumpliendo restricciones y flexibilizando el error para aumentar el rendimiento del algoritmo). En regresión lo que se busca es la línea que contenga mayor número de puntos y que el margen sea lo más reducido posible.
