## Quitando restricciones y aadiendo flexibilidad al modelo

si no fusemos flexibles a la hora de permitir algn error, nos encontaramos con el problema de que el modelo estara muy limitado en cuanto a las posibles soluciones que es capaz de encontrar y, adems, posiblemente dichas soluciones no seran buenas. La idea es permitir que el modelo tenga mayor capacidad para *jugar* con el problema. Esto se logra en un SVM con el hiperparmetro *C*, que permite por un lado que la separacin sea la mayor posible y por el otro, que se produzcan violaciones de las restricciones impuestas al margen.
